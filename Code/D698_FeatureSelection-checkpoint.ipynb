### FEATURE SELECTION Process

```python
# Import modules from the 'scikit-learn' (or 'sklearn') libraries

from sklearn.feature_selection import SelectKBest, chi2, f_classif, f_regression, mutual_info_classif, mutual_info_regression
#imports various feature selection methods and tools from the 'feature_selection' module within 'sklearn'
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
#create ensemble models based on random forests for classification and regression tasks
from sklearn.feature_selection import SelectFromModel
#select features based on a machine learning model's feature importance
from sklearn.preprocessing import LabelEncoder
#encode categorical labels into numerical values
```

```python
def feature_selection(data, target, n_features_to_select=None, problem_type='regression'):
    # Separate categorical and numerical features
    categorical_features = data.select_dtypes(include=['category']).columns.tolist()
    numerical_features = data.select_dtypes(exclude=['category']).columns.tolist()

    # Encode categorical features using LabelEncoder
    data_encoded = data.copy()
    for feature in categorical_features:
        le = LabelEncoder()
        data_encoded[feature] = le.fit_transform(data[feature])

  # Feature selection for numerical features (for regression problem)
    score_func_num = mutual_info_regression
    selector_num = SelectKBest(score_func=score_func_num, k=n_features_to_select)
    selector_num.fit(data_encoded[numerical_features], target)


    # Feature selection for categorical features using Random Forest
    if problem_type == 'classification':
        selector_cat = SelectFromModel(RandomForestClassifier(n_estimators=100))
    else:
        selector_cat = SelectFromModel(RandomForestRegressor(n_estimators=100))
    selector_cat.fit(data_encoded[categorical_features], target)

    # Combine selected features from both numerical and categorical features
    selected_features_num = [numerical_features[i] for i in selector_num.get_support(indices=True)]
    selected_features_cat = [categorical_features[i] for i in selector_cat.get_support(indices=True)]
    selected_features = selected_features_num + selected_features_cat

    return selected_features

# Data usage:
data = cali_df2
target = data['WFIR_RISKS']
selected_features = feature_selection(data, target, n_features_to_select=50, problem_type='regression')
print(selected_features)
```

```python
# Create a DataFrame from the 'selected_features' list
selected_features_df = pd.DataFrame({'Selected Features': selected_features})

# Print the DataFrame
print(selected_features_df)
```
